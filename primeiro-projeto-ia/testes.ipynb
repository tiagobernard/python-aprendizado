```python
import pandas as pd

df = pd.read_csv("toalhas.csv")
df
```

```python
df.plot(kind="scatter", x="quantidade", y="custo")
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["quantidade"]]
y = df[["custo"]]

modelo.fit(x,y)
```

```python
print(modelo.predict([[23]])[0][0])
```
